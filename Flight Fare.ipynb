## Importing the Data

```python
import pandas as pd
```

```python
Fare = pd.read_csv("Data.csv")
```

## Data Analysing

```python
Fare.head()
```

```python
Data_Table = Fare
```

```python
Data_Table['Additional_Info'].unique()
```

```python
Data_Table.head()
```

## Converting Additonal Info to Integer with the help of Level Encoder

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Data_Table['Additional_Info'] = le.fit_transform(Data_Table['Additional_Info'])
```

```python
Data_Table.head()
```

## Replace the String Total Stops to Integer Total Stops

```python
Data_Table['Total_Stops'].unique()
```

```python
Data_Table.replace({'non-stop':0 ,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4},inplace=True)
```

```python
Data_Table.head()
```

```python
Data_Table['Additional_Info'].unique()
```

## Seprating Duration time and Duration min

```python
dur=list(Data_Table['Duration'])
```

```python
dur_hr=[]
dur_min=[]
for i in (dur):
    a=i.split()
    if len(a) == 2:
        dur_hr.append(int(a[0][:-1]))
        dur_min.append(int(a[1][:-1]))
    else:
        if a[0][:-1] == 'm':
            dur_min.append(int(a[0][:-1]))
            dur_hr.append(0)
        else:
            dur_min.append(0)
            dur_hr.append(int(a[0][:-1]))
```

```python
Data_Table['Duration_hr']=dur_hr
```

```python
Data_Table['Duration_min']=dur_min
```

```python
Data_Table.head()
```

```python
Data_Table.drop('Duration',axis=1,inplace=True)
```

```python
Data_Table.head()
```

## Seprating Arival and Departure hr and min 

```python
Data_Table['Arrival_Time_hr'] = pd.to_datetime(Data_Table['Arrival_Time']).dt.hour
```

```python
Data_Table['Arrival_Time_min'] = pd.to_datetime(Data_Table['Arrival_Time']).dt.minute
```

```python
Data_Table['Dep_Time_hr'] = pd.to_datetime(Data_Table['Dep_Time']).dt.hour
```

```python
Data_Table['Dep_Time_min'] = pd.to_datetime(Data_Table['Dep_Time']).dt.hour
```

```python
Data_Table.head()
```

```python
Data_Table.drop('Arrival_Time',axis=1,inplace=True)
```

```python
Data_Table.drop('Dep_Time',axis=1,inplace=True)
```

```python
Data_Table.head()
```

```python
Data_Table.drop('Route',axis=1,inplace=True)
```

```python
Data_Table
```

## Modifing the Date

```python
Data_Table['Day_of_Journey'] = pd.to_datetime(Data_Table['Date_of_Journey'],format="%d/%m/%Y").dt.day
```

```python
Data_Table['Month_of_Journey'] = pd.to_datetime(Data_Table['Date_of_Journey'],format="%d/%m/%Y").dt.month
```

```python
Data_Table['Year_of_Journey'] = pd.to_datetime(Data_Table['Date_of_Journey'],format="%d/%m/%Y").dt.year
```

```python
Data_Table.head()
```

```python
Data_Table.drop('Date_of_Journey',axis=1,inplace=True)
```

## Seprating Airline,Source and Destination using Dummy Variable

```python
des = Data_Table['Destination']
des = pd.get_dummies(des,drop_first=True)
```

```python
des.head()
```

```python
src = Data_Table['Source']
src = pd.get_dummies(src,drop_first=True)
```

```python
src.head()
```

```python
air = Data_Table['Airline']
air = pd.get_dummies(air,drop_first=True)
```

```python
air.head()
```

```python
air.shape
```

```python
Data = pd.concat([Data_Table,des,src,air],axis=1)
```

```python
Data.head()
```

```python
Data.drop('Airline',axis=1,inplace=True)
```

```python
Data.drop('Source',axis=1,inplace=True)
```

```python
Data.drop('Destination',axis=1,inplace=True)
```

```python
Data.head()
```

```python
Data.shape
```

## Correlating the model to check the important value

```python
cr=Data.corr()
cr['Price'].sort_values(ascending=True)
```

## sice Year of journey has no corelation with price we will drop it

```python
Data.drop('Year_of_Journey',axis=1,inplace=True)
```

```python
Data.shape
```

## Spliting train and test set

```python
from sklearn.model_selection import train_test_split
train,test = train_test_split(Data,test_size=0.2,random_state=40)
```

```python
train.shape
```

```python
test.shape
```

```python
features = train.drop('Price',axis=1)
```

```python
labels = train['Price'].copy()
```

## Creating Pipeline

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
pipeline=Pipeline([
    ('imputer',SimpleImputer(strategy="median")),
    ('std',StandardScaler()),
])
```

```python
pipeline_data = pipeline.fit_transform(features)
```

```python
pipeline_data
```

## Creating the model

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import HuberRegressor
from sklearn.linear_model import LogisticRegression

#model = LinearRegression()
#model = DecisionTreeRegressor()
model = RandomForestRegressor()
#model = HuberRegressor()
#model = LogisticRegression()
```

```python
model.fit(pipeline_data,labels)
```

## Checking some train set data

```python
some_data = features[:5]
some_labels = labels[:5]
data_prep = pipeline.transform(some_data)
model.predict(data_prep)
```

```python
list(some_labels)
```

## Finding the error value

```python
from sklearn.metrics import mean_squared_error
import numpy as np
pridction = model.predict(pipeline_data)
mse = mean_squared_error (pridction ,labels)
rmse = np.sqrt(mse)
```

```python
rmse
```

## Applying cross validation

```python
from sklearn.model_selection import cross_val_score
score = cross_val_score(model , pipeline_data , labels , scoring='neg_mean_squared_error',cv=10)
rmse_score = np.sqrt(-score)
```

## Checking the score

```python
def score(rmse):
    print('score = ',rmse)
    print('mean = ',rmse.mean())
    print('std = ',rmse.std())
```

```python
score(rmse_score)
```

## Saving the model

```python
from joblib import dump
dump(model,'Flight Fare.joblib')
```

## Testing the model

```python
X_Test = test.drop('Price',axis=1)
Y_Test = test['Price'].copy()

data_from_pipeline = pipeline.transform(X_Test)

final_pridct = model.predict(data_from_pipeline)

final_mse = mean_squared_error(final_pridct,Y_Test)

final_rmse = np.sqrt(final_mse)

final_rmse

accuracy = model.score(data_from_pipeline,Y_Test)*100

accuracy
```
